```python
filename = "bus1.txt"
f = open("bus1.txt", "r", encoding="utf8")
txtt = f.read()
print(txtt)

print('\n')

txt = txtt.split()
print(txt)
print('\n')

#Number of Alphabets in file "bus1.txt"
print(filename, len(txtt))

print('\n')

#Number of words in file "bus1.txt"
print(filename, len(txt))
```

```python
#Make the dictionary
dictionry= dict()
for words in txt:
    dictionry[words] = dictionry.get(words,0) + 1
#print(dictionry)

print('\n')

#Count the frequency of words in the file
big_words = None
big_count = None
for word,count in dictionry.items():
    print(word,count)
    if big_words is None or count > big_count:
        big_words = word
        big_count = count
        
print('\n')

#Most frequent word in file "bus1.txt"
print(big_words, big_count)
```

```python
#open every file in a folder
import os,glob
folder_path = "Data\\Real"
for filename in glob.glob(os.path.join(folder_path, '*.txt')):
    with open(filename,"r", encoding="utf8") as f:
        text = f.read().split()
        #Make the dictionary
        dictionry = dict()
        for words in text:
            dictionry[words] = dictionry.get(words,0) + 1
        #print(dictionry)

        print('\n')

        #Count the frequency of words in the file
        big_words = None
        big_count = None
        for word,count in dictionry.items():
            print(word,count)
            if big_words is None or count > big_count:
                big_words = word
                big_count = count

        print('\n')

        #Most frequent word in file "bus1.txt"
        print(big_words, big_count)
        print(len(dictionry))
```
