### ROC-curve plots

```python
import re
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


plt.figure(figsize=[5,4])

dicc = {"LogisticRegression": "LR",
            "SVC": "SV",
            "DecisionTreeClassifier": "TC",
            "RandomForestClassifier": "RF",
            "MultinomialNB": "MNB",
        "BernoulliNB": "BNB",
        "AdaBoostClassifier": "AB"}
for i in range(top_by_both_f1_fake_and_real.shape[0]):
    tpr = top_by_both_f1_fake_and_real.iloc[i,]['fpr']
    fpr = top_by_both_f1_fake_and_real.iloc[i,]['tpr']
    auc = top_by_both_f1_fake_and_real.iloc[i,]['test-roc_auc']
    weight_scheme = top_by_both_f1_fake_and_real.iloc[i,]['weight scheme']
    classifier = top_by_both_f1_fake_and_real.iloc[i,]['classifier']
    short_classifier = dicc[classifier]
    c = top_by_both_f1_fake_and_real.iloc[i,]['c']
    w = top_by_both_f1_fake_and_real.iloc[i,]['w']
    f = top_by_both_f1_fake_and_real.iloc[i,]['f']
    cwf = str(c)+"c-"+str(w)+"w-"+str(f)+"f-"
    label = cwf+" "+weight_scheme+" "+short_classifier
    auc =  float(auc)
    def string_to_vector(s):
        s = s.replace("\r", " ")
        s = s.replace("\n", " ")
        s = s.replace("[", " ")
        s = s.replace("]", " ")
        s = s.strip()
        s = re.sub(r"\s+", " ", s)
        s_split = s.split()
        v = []
        for item in s_split:
            v.append(float(item))
        return np.array(v)
    fpr = string_to_vector(fpr)
    tpr = string_to_vector(tpr)
    plt.plot(fpr, tpr, label='{model} ROC (area = {a:0.2f})'.format(model=label,a=auc), alpha=1, linestyle='-')
    
# Custom settings for the plot 
plt.plot([0, 1], [0, 1],'r-', color='blue')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (Real news detected as fake news)')
plt.ylabel('True Positive Rate (Fake News)')
#plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")

plt.savefig('ROC-curves.jpg', dpi=300)

plt.show()


```

## Recall some statistical concepts

![title](statistics\1.png)

Source: 
1. https://stackabuse.com/understanding-roc-curves-with-python/
2. https://machinelearningmastery.com/roc-curves-and-precision-recall-curves-for-classification-in-python/

![title](statistics\2.png)

- Recall: Out of all the positive classes (real news), how many instances were identified correctly.

    * **Recall = TP / (TP + FN)**      

- Precision: Out of all the predicted positive instances (real news), how many were predicted correctly.
    * **Precision = TP / (TP + FP)**

- F-Score: From Precision and Recall, F-Measure is computed and used as metrics sometimes.
- F – Measure is nothing but the harmonic mean of Precision and Recall.
    * F-Score = (2 * Recall * Precision) / (Recall + Precision)
